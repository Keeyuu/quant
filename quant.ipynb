```python
ri = jqdatasdk.get_bars('000001.XSHE', count=1, fields=[
                   'date', 'open', 'close', 'high', 'low', 'volume', 'money'], unit='1d', include_now=True, df=True)
ri
```

```python
ri.iloc[1]
```

```python
# fenxing
def calc_base(org):
    list=[]
    len_ = len(org)
    if len_<= 1:
        return list
    index=1
    while index+1 < len_:
        find_fenx(org,index,list)
        index+=1
    return filter_fenx(list)

def find_fenx(org, index,list):
    if org.iloc[index]['low'] < org.iloc[index-1]['low'] and org.iloc[index]['low'] < org.iloc[index+1]['low']:
        data=dict(org.iloc[index])
        data['type']='high'
        data['index']=index
        list.append(data)
    elif org.iloc[index]['high'] > org.iloc[index-1]['high'] and org.iloc[index]['high'] > org.iloc[index+1]['high']:
        data = dict(org.iloc[index])
        data['type'] = 'low'
        data['index']=index
        list.append(data)


def filter_fenx(list):
    last_obj={'type':'init','index':-99,'high':-99999,'low':99999}
    clean_1=[]
    for i in list:
        if i['type']==last_obj['type']:
            if i['type']=='high' and i['high']>last_obj['high']:
                last_obj=i
            elif i['type'] == 'low' and i['low'] < last_obj['low']:
                last_obj=i
        elif i['index']-last_obj['index']>=3:
            if last_obj['type']!='init':
                clean_1.append(copy.deepcopy(last_obj))
            last_obj=i
    return clean_1

calc_base(ri)
```

```python
for index, row in ri.iterrows():
    print(dict(row))
    break
```
